```python
from matplotlib import pyplot as plt
```

```python
import pandas as pd
```

```python
durations = pd.read_csv('ALS_epoch_durations.csv')
```

```python
durations=durations.set_index('Epoch')
```

```python
durations.mean()
```

```python
fig = plt.figure(1)
ax = fig.add_subplot(111)
plt.plot(durations.mean())
plt.title('Average Duration Epoch SGD')
plt.ylim(0, 5600)
plt.ylabel('time (s)')
ax.set_xticklabels(['1000_users', '5000_users', '10000_users',  '30000_users', '70000_users','138493_users'], rotation = 45, ha="right")
```
